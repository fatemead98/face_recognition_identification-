```python

import face_recognition
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

### load images and convet them to RGB

```python
img1 = face_recognition.load_image_file('/content/drive/MyDrive/download.jpeg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = face_recognition.load_image_file('/content/drive/MyDrive/download1.jpeg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
cv2_imshow(img1)
cv2_imshow(img2)
```

find face location 

```python
face1Loc = face_recognition.face_locations(img1)[0]
face2Loc = face_recognition.face_locations(img2)[0]
```

```python
encode1= face_recognition.face_encodings(img1)[0]
encode2 = face_recognition.face_encodings(img2)[0]
```

```python
pic1=cv2.rectangle(img1,(face1Loc[3],face1Loc[0]),(face1Loc[1],face1Loc[2]),(255,0,255),2)
pic2=cv2.rectangle(img2,(face2Loc[3],face2Loc[0]),(face2Loc[1],face2Loc[2]),(255,0,255),2)
cv2_imshow(pic1)
cv2_imshow(pic2)
```

```python
#result of compare should be false 
results = face_recognition.compare_faces([encode1],encode2)
results 
```

```python
faceDis = face_recognition.face_distance([encode1],encode2)
faceDis
```

```python
a=cv2.putText(img2,f'{results} {round(faceDis[0],2)}',(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)
cv2_imshow(a)
```
